## Background

After running quite a few CNNs on the MIAS dataset, we determined we needed more training data in order to more accurately be able to classify mammogram images as Tumor/No Tumor. One major hurdle is that the majority of labeled mammographic databases are not publicly available. We were, however, able to find the DDSM and CBIS-DDSM datasets that had 55,890 images of both full breast image, mask ROI (region of interest) images, as well as cropped preprocessed images. We proceeded with modeling the already preprocessed images in the interest of time.

### DDSM and CBIS-DDSM Datasets

The CBIS-DDSM (Curated Breast Imaging Subset of DDSM) dataset is an updated and standardized version of the Digital Database for Screening Mammography (DDSM). The images were pre-processed and converted to 299 pixels x 299 pixels by extracting the ROIs (regions of interest) and were stored as tfrecords files for TensorFlow in a [Kaggle Challenge](https://www.kaggle.com/skooch/ddsm-mammography). The dataset contains 55,890 training examples, of which 14% are positive and the remaining 86% negative, divided into 5 tfrecords files. The positive (CBIS-DDSM) images had their ROIs extracted using the masks with a small amount of padding to provide context. Each ROI was then randomly cropped three times into 598x598 images, with random flips and rotations, and then the images were resized down to 299x299.

**label_normal**
- 0 for negative (No Tumor)
- 1 for positive (Tumor)

**label full multi-class labels**
- 0 is negative
- 1 is benign calcification
- 2 is benign mass
- 3 is malignant calcification
- 4 is malignant mass

**Note:**
The data was separated into training and test as per the division in the CBIS-DDSM dataset. The test files have been divided equally into test and validation data. However, the split between test and validation data was done incorrectly, resulting in the test numpy files containing only masses and the validation files containing only calcifications. These files were combined for use in order to have balanced and complete final validation dataset. We used only the training dataset tfrecords files for modeling and used the other fixed validation set for a final check on the model performance and to help confirm there is no data leakage in the training dataset.

## Data Import, Cleaning & EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os, shutil
import cv2
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.callbacks import EarlyStopping
import tensorflow as tf
import glob
from sklearn.preprocessing import LabelEncoder
from tqdm import tqdm
from colorama import Fore, Back, Style
from mpl_toolkits.axes_grid1 import ImageGrid
import random
```

```python
tf.random.set_seed(1234)
```

```python
images=[]
labels=[]

feature_dictionary = {
    'label': tf.io.FixedLenFeature([], tf.int64),
    'label_normal': tf.io.FixedLenFeature([], tf.int64),
    'image': tf.io.FixedLenFeature([], tf.string)
    }

dataset_dir = './data/ddsm/'
```

```python
# Functions for Training data

def _parse_function(example, feature_dictionary=feature_dictionary):
    parsed_example = tf.io.parse_example(example, feature_dictionary)
    return parsed_example

def read_data(filename):
    full_dataset = tf.data.TFRecordDataset(filename,num_parallel_reads=tf.data.experimental.AUTOTUNE)
    full_dataset = full_dataset.shuffle(buffer_size=31000)
    full_dataset = full_dataset.cache()
    print("Size of Training Dataset: ", len(list(full_dataset)))
    
    feature_dictionary = {
    'label': tf.io.FixedLenFeature([], tf.int64),
    'label_normal': tf.io.FixedLenFeature([], tf.int64),
    'image': tf.io.FixedLenFeature([], tf.string)
    }   

    full_dataset = full_dataset.map(_parse_function, num_parallel_calls=tf.data.experimental.AUTOTUNE)
    print(full_dataset)
    for image_features in full_dataset:
        image = image_features['image'].numpy()
        image = tf.io.decode_raw(image_features['image'], tf.uint8)
        image = tf.reshape(image, [299, 299,1])        
        image=image.numpy()
        image=cv2.resize(image,(100,100))
        #image=cv2.resize(image,(299,299))
        #image=cv2.merge([image,image,image])        
        #plt.imshow(image)
        images.append(image)
        labels.append(image_features['label_normal'].numpy())
```

```python
filenames=[f'{dataset_dir}/archive/training10_0/training10_0.tfrecords',
          f'{dataset_dir}/archive/training10_1/training10_1.tfrecords',
          f'{dataset_dir}/archive/training10_2/training10_2.tfrecords',
          f'{dataset_dir}/archive/training10_3/training10_3.tfrecords',
          f'{dataset_dir}/archive/training10_4/training10_4.tfrecords']

for file in filenames:
    read_data(file)
    
print(len(images))
print(len(labels))
```

```python
def show_grid(files, row=3, col=3):
    grid_files = random.sample(files, row*col)
    images = []
    for image_path in tqdm(grid_files):
        #img = cv2.resize(cv2.imread(image_path), dsize=(512,512))
        img = cv2.resize(cv2.imread(image_path), dsize=(256,256))
        images.append(img)

    fig = plt.figure(figsize=(col*5, row*5))
    grid = ImageGrid(fig, 111,  # similar to subplot(111)
                     nrows_ncols=(col, row),  # creates 2x2 grid of axes
                     axes_pad=0.05,  # pad between axes in inch.
                     )

    for ax, im in zip(grid, images):
        # Iterating over the grid returns the Axes.
        ax.imshow(im)
        ax.set_xticks([])
        ax.set_yticks([])
    plt.show()
```
